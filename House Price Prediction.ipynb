```python
t=df.iloc[:, list(range(5)) + [10]]
t
```

```python
t2=df.iloc[0:,0:1]
t2
```

```python
t2.isnull().sum()
```

```python
from sklearn.linear_model import LinearRegression
```

```python
lr=LinearRegression()
```

```python
lr.fit(t,t2)
```

```python
s=lr.coef_[0]
s
```

```python
intercept=lr.intercept_
intercept
```

```python
x=int(input("Enter the area"))
pred=s*x+intercept
pred
```

```python
pred3=lr.predict(t)
pred3
```

```python
from sklearn.metrics import r2_score
print(r2_score(t2,pred3))
```
